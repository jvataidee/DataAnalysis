```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

import zipfile
```

```python
path_to_zip_file = r"/content/archive.zip"
directory_to_extract_to = r"/content"

with zipfile.ZipFile(path_to_zip_file, 'r') as zip_ref:
    zip_ref.extractall(directory_to_extract_to)
```

```python
df = pd.read_csv(r"/content/State of Data 2021 - Dataset - Pgina1.csv")
df
```

```python
df.columns
```

```python
senoridade = df.iloc[:, [0, 9, 10, 11, 12, 14, 15, 16, 17]]
senoridade.head()
```

```python
senoridade["('P2_g ', 'Nivel')"].value_counts()
```

```python
senior  = senoridade[senoridade["('P2_g ', 'Nivel')"] =="Sênior"]
senior.head()
```

```python
senior["('P1_h ', 'Nivel de Ensino')"].value_counts().plot(kind = "bar")
```

```python
quantidade_senior = len(senior)
quantidade_senior
```

```python
ds_final = pd.DataFrame(senior["('P1_h ', 'Nivel de Ensino')"].value_counts())
ds_final["Porcentagem"] = round((ds_final["('P1_h ', 'Nivel de Ensino')"]/quantidade_senior)*100,2)
```

```python
ds_final
```
